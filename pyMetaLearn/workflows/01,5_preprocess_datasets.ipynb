```python
import cPickle
import glob
import os
import time

import arff

import pyMetaLearn.openml.manage_openml_data
from pyMetaLearn.openml.openml_task import OpenMLTask as Task
```

```python
pyMetaLearn.openml.manage_openml_data.set_local_directory(
    "/home/feurerm/thesis/datasets/openml/")
local_directory = pyMetaLearn.openml.manage_openml_data.get_local_directory()
dataset_dir = os.path.join(local_directory, "datasets")
split_dir = os.path.join(local_directory, "splits")
if not os.path.exists(split_dir):
    os.mkdir(split_dir)

dataset_list = glob.glob(os.path.join(dataset_dir, "did*.xml"))
used_datasets = []

dids = []
for dataset_filename in dataset_list:
    did = os.path.split(dataset_filename)[1]
    dids.append(int(did.replace("did", "").replace(".xml", "")))
dids.sort()
```

```python
for did in dids:
    task_file = os.path.join(pyMetaLearn
                .openml.manage_openml_data.get_local_directory(),
                "custom_tasks", "did_%d.pkl" % did)
    
    if not os.path.exists(task_file):
        # print "Skipping dataset %s" % dataset_filename
        continue
    
    with open(task_file) as fh:
        task_args = cPickle.load(fh)
        task = Task(**task_args)
        dataset = pyMetaLearn.openml.manage_openml_data.get_local_dataset(
            task.dataset_id)
        
    # Preprocess the dataset!
    preprocessing_start_time = time.time()
    task.get_dataset_as_pandas()
    X, Y = task.get_dataset()
    preprocessing_end_time = time.time()

    # Calculate the metafeatures of the whole dataset
    metafeatures_start_time = time.time()
    print "Metafeatures for dataset %d: %s" % (did, dataset._name)
    dataset.get_metafeatures()
    metafeatures_end_time = time.time()
    print "took %f seconds" % (metafeatures_end_time - metafeatures_start_time)
    
    for test_fold in range(1):
        print "Calculate the metafeatures for test split %d" % test_fold,
        # Calculate the metafeatures of the three test splits
        train_split, test_split = task._get_fold(X, Y, fold=test_fold, folds=3)
        splits = {}
        splits['description'] = "Preliminary dataset splits generated by Matthias Feurer"
        splits['relation'] = "dataset_%d_%s_splits" % (did, dataset._name)
        splits['attributes'] = [('type', ('TRAIN', 'TEST')),
                                ('rowid', 'INTEGER'),
                                ('repeat', 'INTEGER'),
                                ('fold', 'INTEGER')]
        splits['data'] = [('TRAIN',idx,0,0) for idx in train_split]
        splits['data'].extend([('TEST',idx,0,0) for idx in test_split])
        # Estimation procedure 6, for safety add the properties to the filename
        # From now on, there is only one testfold left
        # More testfolds should now be added to the splits file
        dataset_identifier = "did_%d_%s_" % (did, dataset._name)
        split_identifier = "generated:6:holdout:percentage=33:stratified=True.arff"
        split_file = os.path.join(split_dir, dataset_identifier + split_identifier)
        with open(split_file, "w") as fh:
            arff.dump(splits, fh)
        
        test_split_metafeatures_start_time = time.time()
        dataset.get_metafeatures(split_file_name=split_file)
        test_split_metafeatures_end_time = time.time()
        print "took %f seconds" % (test_split_metafeatures_end_time - test_split_metafeatures_start_time)  
        
        # Calculate the metafeatures of the ten validation splits per test split!
        valid_splits = {}
        valid_splits['description'] = "Preliminary dataset splits generated by Matthias Feurer"
        valid_splits['relation'] = "dataset_%d_%s_splits" % (did, dataset._name)
        valid_splits['attributes'] = [('type', ('TRAIN', 'TEST')),
                                      ('rowid', 'INTEGER'),
                                      ('repeat', 'INTEGER'),
                                      ('fold', 'INTEGER')]
        valid_splits['data'] = []
        
        for fold in range(10):
            valid_train_split, valid_test_split = task._get_fold(X[train_split], Y[train_split], fold=fold, folds=10)
            valid_splits['data'].extend([('TRAIN',idx,0,fold) for idx in valid_train_split])
            valid_splits['data'].extend([('TEST',idx,0,fold) for idx in valid_test_split])
            
        cv_split_identifier = "generated:10foldcrossvalidationinside33percendholdout.arff"
        cv_split_file = os.path.join(split_dir, dataset_identifier + cv_split_identifier)
        with open(cv_split_file, "w") as fh:
            arff.dump(valid_splits, fh)
        
        print "Calculate the cross validation metafeatures", 
        cv_metafeatures_start_time = time.time()
        dataset.get_metafeatures(split_file_name=cv_split_file)
        cv_metafeatures_end_time = time.time()
        print "took %f seconds" % (cv_metafeatures_end_time - cv_metafeatures_start_time) 
```
