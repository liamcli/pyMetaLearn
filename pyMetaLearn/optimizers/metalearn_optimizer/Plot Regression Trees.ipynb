```python
import numpy as np
import scipy
import sklearn.tree
import sklearn.svm
import cPickle
```

```python
input_file = "/home/feurerm/thesis/experiments/2014_04_25_predict_distances/testfold_0-3_spearman_rank_perm.pkl"
with open(input_file) as fh:
    X, Y, metafeatures = cPickle.load(fh) 
```

```python
loo_mae = []
loo_rho = [] 
models = []

for idx in range(metafeatures.shape[0]):
    leave_out_dataset = metafeatures.index[idx]

    train = []
    valid = []
    for cross in X.index:
        if leave_out_dataset in cross:
            valid.append(cross)
        else:
            train.append(cross)

    X_train = X.loc[train].values
    Y_train = Y.loc[train].values
    X_valid = X.loc[valid].values
    Y_valid = Y.loc[valid].values

    model = sklearn.tree.DecisionTreeRegressor(max_depth=5)
    model.fit(X_train, Y_train)
    models.append(model)
    predictions = model.predict(X_valid)
    rho =  scipy.stats.kendalltau(Y_valid, predictions)[0]
    mae = sklearn.metrics.mean_absolute_error(predictions, Y_valid)

    loo_mae.append(mae)
    loo_rho.append(rho)
```

```python
mae = np.mean(loo_mae)
mae_std = np.std(loo_mae)
rho = np.mean(loo_rho)
rho_std = np.std(loo_rho)

for idx in range(metafeatures.shape[0]):
    print "%20s %5f %5f" % (metafeatures.index[idx], loo_mae[idx], loo_rho[idx])

print "MAE", mae, mae_std
print "Mean tau", rho, rho_std 
```

```python
import StringIO
import pydot
```

```python
for idx in range(metafeatures.shape[0]):
    dot_data = StringIO.StringIO()
    sklearn.tree.export_graphviz(models[idx], out_file=dot_data)
    graph = pydot.graph_from_dot_data(dot_data.getvalue()) 
    graph.write_pdf("%s.pdf" % metafeatures.index[idx]) 
```
